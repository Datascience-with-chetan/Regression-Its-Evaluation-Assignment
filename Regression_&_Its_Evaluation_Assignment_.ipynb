1.What is Simple Linear Regression?
- Simple linear regression is a statistical method used to model and quantify the relationship between two continuous (quantitative) variables: one independent variable (also called a predictor or explanatory variable) and one dependent variable (also called a response or outcome variable).

- The model assumes a linear relationship, meaning that changes in the independent variable correspond to proportional changes in the dependent variable.
- Simple linear regression only uses one independent variable.

- The main objectives are to predict the value of the dependent variable based on the independent variable and to assess the strength and direction of their relationship.

- The best-fit line is found by minimizing the sum of squared differences (errors) between observed and predicted values—a method known as ordinary least squares.

- It assumes linearity, independence, constant variance (homoscedasticity), and normally distributed errors.

- Applications:
  - Predicting outcomes (like predicting exam scores from study hours) or understanding relationships (such as how temperature affects crop yield).

  - If there is more than one independent variable, the method is called multiple linear regression



2.What are the key assumptions of Simple Linear Regression?
- The key assumptions of Simple Linear Regression are:

- Linearity: The relationship between the independent variable (X) and the dependent variable (Y) is linear, meaning changes in X produce proportional changes in Y.

- Independence: The residuals (differences between observed and predicted Y) are independent of each other. This means the error associated with one observation does not influence another.

- Homoscedasticity: The residuals have constant variance across all levels of the independent variable. In other words, the spread of errors should be roughly the same for all predicted values of Y.

- Normality of Residuals: The residuals should be (approximately) normally distributed. This is particularly important when using regression for inference such as confidence intervals and hypothesis testing.

- No significant outliers: Outliers can distort the relationship and violate normality and homoscedasticity assumptions.

- Continuous variables: Both X and Y should be continuous (quantitative) variables.

- In the case of multiple regression (not simple regression), no multicollinearity among predictors is also required, but for simple linear regression with only one predictor, this is not applicable.

- If these assumptions are violated, the results and interpretations of the regression analysis may not be valid. Indications of violations can often be detected using residual plots and appropriate statistical tests

3.What is heteroscedasticity, and why is it important to address in regression models?
- Heteroscedasticity in regression analysis refers to a situation where the variance of the residuals (errors) is not constant across all levels of the independent variable(s). In simpler terms, as the value of your predictor changes, the spread (scatter) of the regression errors increases or decreases, rather than remaining uniform.

- This condition violates a key assumption of linear regression models, which expect homoscedasticity (constant variance of residuals). Heteroscedasticity can often be detected visually by plotting residuals versus predicted values—if you observe a pattern where errors “fan out” or “narrow in,” heteroscedasticity is likely present.

  - Addressing heteroscedasticity is important because:

- Standard errors become unreliable: The usual methods for estimating the precision of regression coefficients assume constant variance. When this is violated, standard errors can be biased, leading to unreliable confidence intervals and hypothesis tests.

- Inference may be invalid: Even if coefficient estimates remain unbiased, statistical significance (p-values) will be inaccurate, potentially leading to wrong conclusions.

- Model efficiency decreases: Ordinary Least Squares (OLS) estimates become less efficient—they are still unbiased but are not optimal, increasing the risk of drawing misleading results or missing important predictors.

- Heteroscedasticity does not bias the regression coefficients themselves, but it undermines the trustworthiness of any inferential statistics you report, such as standard errors, t-values, or R-squared. Therefore, detecting and correcting heteroscedasticity (using robust standard errors, transformation of variables, or alternative modeling techniques) is critical for valid regression analysis

4.What is Multiple Linear Regression?
- Multiple Linear Regression (MLR) is a statistical method used to model the relationship between one dependent variable and two or more independent (explanatory) variables. Unlike simple linear regression, which involves just one predictor, MLR allows for analyzing and predicting outcomes based on the combined influence of several factors
- MLR helps estimate the effect size and significance of each predictor when considering the others.
- The technique is widely used for forecasting, understanding the influence of multiple variables, and in inferential statistics across fields like economics, finance, biology, and more.

- The model assumes that the relationship between dependent and independent variables is linear, the predictors are not excessively correlated with each other (no high multicollinearity), and the residuals have constant variance (homoscedasticity) and are normally distributed.

- Typical applications include measuring how factors such as study hours, prior knowledge, and sleep affect exam performance, or how rainfall, temperature, and fertilizer use together impact crop yields.

- multiple linear regression extends the concepts of simple linear regression to analyze how several factors together affect a single outcome, enhancing predictive power and analytical insight

5.What is polynomial regression, and how does it differ from linear regression?
- Polynomial Regression is a type of regression analysis used to model the relationship between an independent variable and a dependent variable when that relationship is not simply a straight line but exhibits a curve or some type of non-linear pattern. Instead of fitting just a straight line through the data (as in linear regression), polynomial regression fits a curved line — a polynomial function — that can bend and flex to better follow the data points.
- How is Polynomial Regression Different from Linear Regression?

  - Form of Relationship:
Linear regression assumes a straight-line relationship between the predictor and the outcome. This means any change in

y. On the other hand, polynomial regression allows the relationship to bend and curve by including powers of x, enabling it to capture more complex, nonlinear trends in the data.

  - Flexibility:
Because polynomial regression can model curves as well as straight lines, it is more flexible and can better fit data that does not align well to a straight line. However, with greater flexibility comes the risk of overfitting — particularly if very high-degree polynomials are used — where the model captures random noise rather than the true underlying pattern.

  - Interpretation:

```python
#6.Implement a Python program to fit a Simple Linear Regression model to the following sample data: ● X = [1, 2, 3, 4, 5] ● Y = [2.1, 4.3, 6.1, 7.9, 10.2] Plot the regression line over the data points.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape((-1, 1))  # Reshape for sklearn which expects 2D array for features
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, Y)

# Get the predicted values
Y_pred = model.predict(X)

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Data points')

# Plot the regression line
plt.plot(X, Y_pred, color='red', label='Regression line')

# Add title and labels
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')

# Show legend
plt.legend()

# Show the plot
plt.show()

# Print the coefficients
print(f"Intercept (b0): {model.intercept_}")
print(f"Slope (b1): {model.coef_[0]}")
```

```python
#7.Fit a Multiple Linear Regression model on this sample data: ● Area = [1200, 1500, 1800, 2000] ● Rooms = [2, 3, 3, 4] ● Price = [250000, 300000, 320000, 370000]
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Prepare the data
Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

# Combine Area and Rooms as independent variables (features)
X = np.column_stack((Area, Rooms))

# Dependent variable
y = Price

# Create a Multiple Linear Regression model
model = LinearRegression()

# Fit the model on the data
model.fit(X, y)

# Print the coefficients and intercept
print(f"Intercept (b0): {model.intercept_}")
print(f"Coefficients (b1 for Area, b2 for Rooms): {model.coef_}")

# Predict prices based on the model
y_pred = model.predict(X)

# Plotting the actual vs predicted prices for visualization (1D approx)
plt.scatter(range(len(y)), y, color='blue', label='Actual Price')
plt.scatter(range(len(y_pred)), y_pred, color='red', label='Predicted Price')
plt.xlabel('Sample Index')
plt.ylabel('Price')
plt.title('Multiple Linear Regression: Actual vs Predicted Prices')
plt.legend()
plt.show()
```

```python
#8.Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] 3 ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # reshape to 2D array for sklearn
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values using the model
Y_pred = model.predict(X_poly)

# Plotting
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='2nd Degree Polynomial Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.show()

# Print the model coefficients
print(f"Intercept (b0): {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
#9.Create a residuals plot for a regression model trained on this data: ● X = [10, 20, 30, 40, 50] ● Y = [15, 35, 40, 50, 65] Assess heteroscedasticity by examining the spread of residuals
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('X')
plt.ylabel('Residuals (Actual - Predicted)')
plt.title('Residuals Plot')

plt.show()

'''How to assess heteroscedasticity from this plot:
If the residuals appear randomly scattered with roughly constant spread (variance) along all values of X the assumption of homoscedasticity holds.

If the residuals “fan out” or “funnel in” (i.e., the variance of residuals changes systematically with X ) it indicates heteroscedasticity which violates a key regression assumption and can affect inference reliability.'''
```

10.Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.
- As a data scientist facing heteroscedasticity and multicollinearity issues in a real estate house price prediction model using features like area, number of rooms, and location, here are the key steps you should take to address these problems and build a robust regression model:

1. Addressing Heteroscedasticity
Heteroscedasticity means the variance of residuals (errors) varies with the level of predictors, violating the assumption of constant error variance, which can lead to unreliable standard errors and inference.

- Steps to address it:

  - Visualize Residuals: Plot residuals vs. fitted values to confirm heteroscedasticity visually.

  - Transform the Dependent Variable: Apply transformations such as log, square root, or Box-Cox to stabilize variance. For example, predicting log(Price) instead of Price often helps.

  - Use Robust Standard Errors: Instead of classic OLS standard errors, use heteroscedasticity-consistent (robust) standard errors (e.g., White’s standard errors).

  - Weighted Least Squares (WLS): Assign weights inversely proportional to the variance of residuals, giving less weight to points with high variance.

  - Model Alternative Specifications: Consider nonlinear models or generalized linear models if heteroscedasticity persists.

2. Handling Multicollinearity
Multicollinearity occurs when independent variables (e.g., area, rooms, location encoded features) are highly correlated, destabilizing coefficient estimates and making it hard to interpret individual predictors.

- Steps to handle multicollinearity:

  - Detect Multicollinearity:

    - Calculate the Variance Inflation Factor (VIF) for each predictor.

    - High VIF values (commonly > 5 or 10) indicate troublesome multicollinearity.

  - Remove or Combine Variables:

    - Drop or merge highly correlated features (e.g., combine rooms and area into a composite metric).

  - Dimensionality Reduction:

    - Use Principal Component Analysis (PCA) or other dimensionality reduction techniques to create uncorrelated components.

  - Regularization:

    - Use regression techniques like Ridge regression (L2 regularization) or Lasso regression (L1 regularization) to penalize large coefficients and reduce multicollinearity impact.

3. General Recommendations for a Robust Model
  - Feature Engineering for Location:

    - Encode location effectively (e.g., one-hot encoding for neighborhoods or use geospatial clustering).

    - Consider interaction terms if meaningful (e.g., area * location).

  - Check Model Assumptions:

    - After corrections, verify residuals are homoscedastic and roughly normal.

    - Check independence of residuals.

  - Model Validation:

    - Split data into training and testing sets.

    - Use cross-validation to assess model stability.

  - Alternative Models:

    - If linear assumptions are too restrictive, explore tree-based models (Random Forests, Gradient Boosting) which are less sensitive to heteroscedasticity and multicollinearity issues.
